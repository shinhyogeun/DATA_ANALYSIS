### 날짜 전처리 코드

```python
# 날짜 전처리 코드
when = df_1107["날짜"]
def change_date_time(when) :
    for i in range(len(when)):
            listi = list(when[i])
            if listi[0] == "최" or listi[2] == "시" or listi[0] == "실":
                #전처리가 필요한 DATA

                #최초공개 SYTLE
                if listi[0] == "최":
                    when[i] = str(''.join(listi[7:]))

                # 00시간 전 STYLE
                if listi[2] == "시":
                    if 23 - int(''.join(listi[:2])) >= 0:
                        when[i] = '2020. 11. 2.'
                    else:
                        when[i] = '2020. 11. 1.'

                # 실시간 SYTLE
                if listi[4] == "스":
                    when[i] = str(''.join(listi[14:]))
    df_1107["날짜"]=when
change_date_time(when)
```

### 좋아요수,싫어요수

```python
# 좋아요수 전처리
#like=df_1107["좋아요수"]
def change_list_to_number(like):
    for i in range(len(like)):
        listi = like[i]
        before = []
        after = 0
        if len(listi) >= 10:
            where = listi.index("\r")
            before = listi[:where]
            if before == "좋아요": before = ["0"]
        else:
            before = listi
            
        if "만" in before:
            after = int(float(''.join(before[:before.index("만")])) * 10000)
        elif "천" in before:
            after = int(float(''.join(before[:before.index("천")])) * 1000)
        else:
            after = int(float(''.join(before)))
        like[i] = after
        #df_1107["좋아요수"]=like

change_list_to_number(["좋아요\rasdasdasdasda","1만"])
```

```python
# 싫어요수 전처리
like=df_1107["싫어요수"]
def change_list_to_number(like):
    for i in range(len(like)):
        listi = like[i]
        before = []
        after = 0
        
        if len(listi) >= 10:
            where = listi.index("\r")
            before = listi[:where]
            if before == "싫어요": before = ["0"]
        else:
            before = listi
        if "만" in before:
            after = int(float(''.join(before[:before.index("만")])) * 10000)
        elif "천" in before:
            after = int(float(''.join(before[:before.index("천")])) * 1000)
        else:
            after = int(float(''.join(before)))
        like[i] = after
        df_1107["싫어요수"]=like

change_list_to_number(like)
```

## 구독자수

```python
subs=df_1107["구독자수"]

def subs_to_num(subs):
    sub = list(map(float,subs))
    for i in range(len(sub)):
        if sub[i] < 100:
            sub[i] *= 10000
    sub = list(map(int,sub))
    df_1107["구독자수"] = sub
subs_to_num(subs)
```
